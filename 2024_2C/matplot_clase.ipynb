En gráficos separados:

a) Graficar las ventas en función de los años usando gráfico de línea y gráfico de barras horizontal. ¿Cuál te parece que es mejor para este caso?

b) Graficar las ganancias en función de los años usando gráfico de línea y gráfico de barras. ¿Cuál te parece que es el mejor gráfico para este caso?

c) Graficar la cantidad de ventas en el año 2024 en función de los productos usando gráfico de barras y gráfico de torta. ¿Cuál te parece que es el mejor gráfico para este caso?


````python
import matplotlib.pyplot as plt
import numpy as np

años = np.arange(2012, 2022) # Mostrar aparte, es un range, pero devuelve un arreglo de numpy

ventas = [140, 60, 120, 250, 200, 180, 100, 300, 120, 160]
ganancias = [14000, 45000, 20400, 100000, 50000, 25000, 100000, 30000, 15000, 35000]
productos = ["prod_1", "prod_2", "prod_3", "prod_4", "prod_5", "prod_6", "prod_7", "prod_8", "prod_9", "prod_10"]
cant_ventas_productos_2024 = [180, 160, 190, 140, 100, 200, 120, 130, 150, 170]
````

````python
# Gráfico de línea
fig, ax = plt.subplots()

ax.plot(años, ventas) # Default, line plot
ax.set_title('Ventas en función de los años (Gráfico de línea)')
ax.set_xlabel('Años') # Le pone que es el eje x
ax.set_ylabel('Ventas')
ax.set_xticks(años) # Cada barrita del eje x le dice que es, si no se pone, va a ser abitrario que valores muestra
ax.grid()
plt.show()


# # Acá se puede ver la diferencia entre usar 'subplots' (requiere trabajar sobre "ax") y 'figure' (trabajamos todo con "plt")

# Gráfico de barras horizontal
plt.figure()
plt.barh(años, ventas) # bar horizontal
plt.title('Ventas en función de los años (Gráfico de barras)')
plt.xlabel('Años')
plt.ylabel('Ventas')
plt.grid()
plt.show()
````

````python
# Gráfico de línea
plt.figure()
plt.plot(años, ganancias)
plt.title('Ganancias en función de los años (Gráfico de línea)')
plt.xlabel('Años')
plt.ylabel('Ganancias')
plt.grid()
plt.show()

# Gráfico de barras
plt.figure()
plt.bar(años, ganancias)
plt.title('Ganancias en función de los años (Gráfico de barras horizontales)')
plt.xlabel('Ganancias')
plt.ylabel('Años')
plt.grid()
plt.show()
````

````python
# Gráfico de barras
plt.figure()
plt.bar(productos, cant_ventas_productos_2024)
plt.title('Cantidad de ventas en 2024 (Gráfico de barras)')
plt.xlabel('Productos')
plt.ylabel('Cantidad de Ventas')
plt.grid()
plt.show()

# ventas = [140, 60, 120, 250, 200, 180, 100, 300, 120, 160]
# ganancias = [14000, 45000, 20400, 100000, 50000, 25000, 100000, 30000, 15000, 35000]
# productos = ["prod_1", "prod_2", "prod_3", "prod_4", "prod_5", "prod_6", "prod_7", "prod_8", "prod_9", "prod_10"]
# cant_ventas_productos_2024 = [180, 160, 190, 140, 100, 200, 120, 130, 150, 170]

# Gráfico de torta
plt.figure()
plt.pie(cant_ventas_productos_2024, autopct='%1.1f%%', labels=productos)
# %<cant_minima_de_decimales_antes_del_punto>.<cant_max_decimales_dsps_del_punto><type> y dsps le pongo %% para que me agregue
# el % al final dsps del numero
plt.title('Cantidad de ventas en 2024 (Gráfico de torta)')
plt.show()
````
